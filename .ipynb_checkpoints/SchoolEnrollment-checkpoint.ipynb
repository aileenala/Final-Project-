```python
import pandas as pd
import matplotlib.pyplot as plt  

enrollment = pd.read_csv("EnrollmentByNeighborhood2020-2021.csv", sep= ",")

del enrollment['african_american']
del enrollment['american_indian']
del enrollment['asian__not_pacific_islander']
del enrollment['hispanic']
del enrollment['multiracial']
del enrollment['native_hawaiian_or_other_pacific_islander']
del enrollment['white']
del enrollment['student_count_redactions']
del enrollment['attending_feeder']
del enrollment['attending_nonfeeder']
del enrollment['total_with_feeder_status']
del enrollment['feeder_count_redactions']
del enrollment['_id']
del enrollment['grade_level']


query_mask = enrollment['total_students_enrolled'] > 200
schools = enrollment[query_mask]

bargraph = schools.plot.bar()
```
